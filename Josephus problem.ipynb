# 约瑟夫环问题
----
## 问题描述： 
**编号为1 - N的N个士兵围坐在一起形成一个圆圈，从编号为1的士兵开始依次报数（1,2,3...依次报数），数到m的士兵会被杀死出列， 之后的士兵再从1开始报数。直到最后剩下一士兵，求这个士兵的编号。**

## 1. 环形链表
### 时间复杂度：O(m*n)

```python
class Node():
    def __init__(self,data):
        self.data = data
        self.next = None
        
def creatLinkedList(n):
    head = Node(1)
    cur = head
    for i in range(2, n + 1):
        cur.next = Node(i)
        cur = cur.next
    cur.next = head
    return head

def LastSoilder(n, m):
    head = creatLinkedList(n)
    cur = head
    while cur.next != cur:
        cur.next.next = cur.next.next.next
        cur = cur.next.next
    return cur.data
```

```python
LastSoilder(5, 3)
```

## 2. 递归！！！！！
### 时间复杂度: O(n)
----
### 思路：

**N个人的约瑟夫环最终存活的人 = N-1个人的约瑟夫环最终存活的人 = …… = 一个人的约瑟夫环最终存活的人，因此需要知道幸存者每轮的编号变化**  
**以n = 11， m = 3为例，最终存活的人编号为7，即：(图中黄色为每轮死去的人，红色为最终的幸存者）**  
**LastSoilder(11, 3) = 7**  
**LastSoilder(10, 3) = 4**  
**LastSoilder( 9, 3) = 1**  
**...........**  
**LastSoilder( 2, 3) = 2**  
**LastSoilder( 1, 3) = 1**  

**可以看出**：  
**(LastSoilder( n - 1, 3) + m - 1) % n + 1 = LastSoilder( n, 3)，从而可以得到递推公式。**

```python
def LastSoilder(n, m):
    return 1 if n == 1 else (LastSoilder(n - 1, m) + m - 1) % n + 1
```

```python
LastSoilder(4, 3)
```
